```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

df=pd.read_csv("census-income.csv")
# Rename the last column as 'Annual Income'
df.rename(columns={'annual_income': 'Annual Income'}, inplace=True)

# Remove missing values
df.dropna(inplace=True)

# Change labels of categorical data into numerical data using LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

# Split the dataset into train and test sets (70:30) with random state 0
X = df.drop('Annual Income', axis=1)
y = df['Annual Income']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Build a Logistic Regression Model
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
from sklearn.metrics import classification_report

# Make predictions on the test data
y_pred = logistic_model.predict(X_test)

# Get the classification report
classification_report = classification_report(y_test, y_pred, output_dict=True)

# Extract the support value for each class
support_value = classification_report['macro avg']['support']

# Print the total support value
print("Total Support Value:", support_value)

```

```python
accuracy = logistic_model.score(X_test, y_test)

# Print the accuracy score
print("Accuracy Score:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
# Make predictions on the test data
y_pred = logistic_model.predict(X_test)

# Compute the confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

# Calculate the specificity
specificity = tn / (tn + fp)

# Print the specificity
print("Specificity:", specificity)
```

```python
from sklearn.metrics import precision_score

# Make predictions on the test data
y_pred = logistic_model.predict(X_test)

# Calculate the precision when the target is False
precision = precision_score(y_test, y_pred, pos_label="<=50K")

# Print the precision
print("Precision when the target is False:", precision)
```
